```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
stats = pd.read_csv('ps.csv',header = None)
```

```python
stats
```

```python
mem_usages_kitsune = stats[1]
```

```python
total_time_python = 2.81
total_time_go = 1.52
total_time_kitsune = 5000
```

```python
cpu_usages_python = [35.2,45.1,100,100,83.9,86.8,71.3,85.4]

time_variable_python = np.linspace(0,total_time_python,len(cpu_usages_python))
```

```python
cpu_usages_go = [29.4,28.8,31.4,30.4,31.4,32.9,30.4,31.8]

time_variable_go = np.linspace(0,total_time_go,len(cpu_usages_go))
```

```python
cpu_usages_kitsune = stats[0]

time_variable_kitsune = np.linspace(0,total_time_kitsune,len(cpu_usages_kitsune))
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(y=cpu_usages_python,x=time_variable_python,label = 'Python')
sns.lineplot(y=cpu_usages_go,x=time_variable_go,label = 'GoLang')
sns.lineplot(y=cpu_usages_kitsune,x=time_variable_kitsune,label = 'Kitsune')

plt.title('System CPU Usage Over Time for Prediction of 100.000 samples')
plt.ylim(0, 101)
plt.xlabel('Time (s)')
plt.xscale('log')
plt.ylabel('CPU Usage (%)')
plt.show()
```
