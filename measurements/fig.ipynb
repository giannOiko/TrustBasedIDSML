```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
total_time_python = 2.81
total_time_go = 1.52
```

```python
cpu_usages_python = [35.2,45.1,100,100,83.9,86.8,71.3,85.4]

time_variable_python = np.linspace(0,total_time_python,len(cpu_usages_python))
```

```python
cpu_usages_go = [29.4,28.8,31.4,30.4,31.4,32.9,30.4,31.8]

time_variable_go = np.linspace(0,total_time_go,len(cpu_usages_go))
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(y=cpu_usages_python,x=time_variable_python,label = 'Python')
sns.lineplot(y=cpu_usages_go,x=time_variable_go,label = 'GoLang')
plt.title('System CPU Usage Over Time for Prediction of 100.000 samples by LGBM Classifier ')
plt.ylim(0, 101)
plt.xlabel('Time (s)')
plt.ylabel('CPU Usage (%)')
plt.show()
```

```python
mem_usages_python = [20.85,21.85,21.77,21.80,21.80,21.82,21.82]

time_variable_python = np.linspace(0,total_time_python,len(mem_usages_python))
```

```python
mem_usages_go = [20.35,20.37,20.37,20.40,20.40,20.43,20.43,20.43]

time_variable_go = np.linspace(0,total_time_go,len(mem_usages_go))
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(y=mem_usages_python,x=time_variable_python,label = 'Python')
sns.lineplot(y=mem_usages_go,x=time_variable_go,label = 'GoLang')
plt.title('System MEM Usage Over Time for Prediction of 100.000 samples by LGBM Classifier ')

plt.ylim(15, 25)
plt.xlabel('Time (s)')
plt.ylabel('MEM Usage (%)')
plt.show()
```
